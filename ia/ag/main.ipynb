```python
from rainhas import FactoryRainhas
from ag import AG
from roletas import Roleta
```

```python
qtd_genes = 8
tx_mutacao = 0.3
factory = FactoryRainhas(qtd_genes, tx_mutacao)

qtd_populacao = 20
elite = 8 
geracoes = 1000
ag = AG(factory, qtd_populacao, elite, "min", Roleta.minimo_com_seletor)

melhor_individuo = ag.executar(geracoes)
```

```python
print("Melhor indivíduo encontrado:", melhor_individuo._genes)
print("Fitness do melhor indivíduo:", melhor_individuo.fitness())
```
