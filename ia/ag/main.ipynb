```python
from rainhas import FactoryRainhas
from ag import AG
from roletas import Roleta
```

```python
qtd_genes = 4
tx_mutacao = 0.2
factory = FactoryRainhas(qtd_genes, tx_mutacao)

qtd_populacao = 40
elite = int(0.3*qtd_populacao)
geracoes = 1000
ag = AG(factory=factory, 
        n=qtd_populacao, 
        elite=elite, 
        type="min",
        # plato_increase=0.5,
        # plato_size=int(0.25 * geracoes),
        tx_mutacao=tx_mutacao,
        funcao_roleta=Roleta.simples_minimo)

ag.executar(geracoes)
```
