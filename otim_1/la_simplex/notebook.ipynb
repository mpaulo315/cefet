### Exercício 1

***(F) - O método simplex não funciona para problemas de duas variáveis.***

> O método funciona para n variáveis, com n $\geq$ 2.

***(F) - O método simplex sempre encontra uma solução viável.***

> O método só encontra solução viável se houver uma região viável limitada e não vazia.

***(F) - As variáveis básicas são equivalentes às variáveis de folga.***

> As variáveis de folga desempenham papel de ajuste das inequações, transformando as em equações.

***(F) - As variáveis de folga representam custos adicionais no problema.***

> As variáveis de folga não afetam a função objetivo, uma vez que possuem coeficiente nulo.

***(F) - As variáveis de folga são descartadas em uma solução viável.***

> As varáveis de folga são úteis durante a iteração no método Simplex, mas não impactam a interpretação final dos resultados.


### Exercício 3


|  Seção |  A | B  | C  |  Horas totais |
|---|---|---|---|---|
| Corte  |  $8$ | $5$  | $2$  | $3 \times 40 = 120$  |
|  Dobra | $5$  | $10$  |  $4$ |  $10 \times 40 = 100$ |
| Empacotamento | $0.7$ | $1$ | $2$ | $2 \times 40 = 80$ |

Maximizar:
$$Z = A + 1{,}5B + 2C$$

 
Sujeito a:
\begin{align*}
8A + 5B + 2C &\leq 120 \quad (\text{Corte}) \\
5A + 10B + 4C &\leq 400 \quad (\text{Dobra}) \\
0{,}7A + B + 2C &\leq 80 \quad (\text{Empacotamento}) \\
A, B, C &\geq 0
\end{align*}

Adicionando folgas $s_1, s_2, s_3$:
 
\begin{align*}
8A + 5B + 2C + s_1 &= 120 \\
5A + 10B + 4C + s_2 &= 400 \\
0{,}7A + B + 2C + s_3 &= 80
\end{align*}


\begin{array}{c|cccccc|c}
\text{Base} & A & B & C & s_1 & s_2 & s_3 & b \\ \hline
s_1 & 8 & 5 & 2 & 1 & 0 & 0 & 120 \\
s_2 & 5 & 10 & 4 & 0 & 1 & 0 & 400 \\
s_3 & 0{,}7 & 1 & \mathbf{2} & 0 & 0 & 1 & 80 \\ \hline
Z & -1 & -1{,}5 & \mathbf{-2} & 0 & 0 & 0 & 0
\end{array}

 
Variável que entra: $C$ (coeficiente $-2$, o mais negativo).
 
\medskip
Teste da razão mínima:
$$
\frac{120}{2} = 60, \quad \frac{400}{4} = 100, \quad \frac{80}{2} = \mathbf{40} \leftarrow \text{mínimo}
$$
 
Variável que sai: $s_3$. Pivô $= 2$.
 
Nova linha $C = \dfrac{1}{2}\, \text{linha}\, s_3$:
$$
(0{,}35,\ 0{,}5,\ 1,\ 0,\ 0,\ 0{,}5\ |\ 40)
$$
 
Atualizando as demais linhas:
\begin{align*}
\text{Nova } s_1 &= s_1 - 2C: \quad (7{,}3,\ 4,\ 0,\ 1,\ 0,\ -1\ |\ 40) \\
\text{Nova } s_2 &= s_2 - 4C: \quad (3{,}6,\ 8,\ 0,\ 0,\ 1,\ -2\ |\ 240) \\
\text{Nova } Z &= Z + 2C: \quad (-0{,}3,\ -0{,}5,\ 0,\ 0,\ 0,\ 1\ |\ 80)
\end{align*}
 

\begin{array}{c|cccccc|c}
\text{Base} & A & B & C & s_1 & s_2 & s_3 & b \\ \hline
s_1 & 7{,}3 & 4 & 0 & 1 & 0 & -1 & 40 \\
s_2 & 3{,}6 & 8 & 0 & 0 & 1 & -2 & 240 \\
C   & 0{,}35 & 0{,}5 & 1 & 0 & 0 & 0{,}5 & 40 \\ \hline
Z & -0{,}3 & \mathbf{-0{,}5} & 0 & 0 & 0 & 1 & 80
\end{array}


 
Variável que entra: $B$ (coeficiente $-0{,}5$).
 
Teste da razão mínima:
$$
\frac{40}{4} = \mathbf{10} \leftarrow \text{mínimo}, \quad \frac{240}{8} = 30, \quad \frac{40}{0{,}5} = 80
$$ 

Variável que \textbf{sai}: $s_1$. Pivô $= 4$.
 

Nova linha $B = \dfrac{1}{4}\, \text{linha}\, s_1$:
$$
(1{,}825,\ 1,\ 0,\ 0{,}25,\ 0,\ -0{,}25\ |\ 10)
$$
 
Atualizando:
\begin{align*}
\text{Nova } s_2 &= s_2 - 8B: \quad (-11{,}0,\ 0,\ 0,\ -2,\ 1,\ 0\ |\ 160) \\
\text{Nova } C   &= C - 0{,}5B: \quad (-0{,}5625,\ 0,\ 1,\ -0{,}125,\ 0,\ 0{,}625\ |\ 35) \\
\text{Nova } Z   &= Z + 0{,}5B: \quad (0{,}6125,\ 0,\ 0,\ 0{,}125,\ 0,\ 0{,}875\ |\ 85)
\end{align*}
 
\begin{array}{c|cccccc|c}
\text{Base} & A & B & C & s_1 & s_2 & s_3 & b \\ \hline
B   & 1{,}825   & 1 & 0 & 0{,}25  & 0 & -0{,}25 & 10 \\
s_2 & -11{,}0   & 0 & 0 & -2      & 1 & 0       & 160 \\
C   & -0{,}5625 & 0 & 1 & -0{,}125 & 0 & 0{,}625 & 35 \\ \hline
Z   & 0{,}6125  & 0 & 0 & 0{,}125 & 0 & 0{,}875 & 85
\end{array}

 
Todos os coeficientes na linha $Z$ são $\geq 0$: \textbf{solução ótima!}
 
\subsection*{Resultado}
 
$$
\boxed{A = 0, \quad B = 10 \text{ unid.}, \quad C = 35 \text{ unid.}, \quad Z = R\$\ 85{,}00/\text{semana}}
$$
 


### Exercício 4


As variáveis representam a produção em milhares de peças por forno:
*  $x_1$ = blocos de 6 furos (mil unid.)
*  $x_2$ = blocos de 8 furos (mil unid.)
*  $x_3$ = lajotas (mil unid.)

Maximizar:
$$Z = 9x_1 + 10x_2 + 15x_3$$
 
Restrições:
 
1. Capacidade do forno (12 mil peças):
    $$
    x_1 + x_2 + x_3 \leq 12
    $$
    
2. Disponibilidade de argila ($18\ \text{m}^3 = 18{.}000\ \text{dm}^3$ por forno).
    Consumo por unidade: $1{,}2$, $1{,}6$ e $1{,}8\ \text{dm}^3$. Em milhares de peças:
    $$
    1200x_1 + 1600x_2 + 1800x_3 \leq 18{.}000
    $$
    Dividindo por 200:
    $$
    6x_1 + 8x_2 + 9x_3 \leq 90
    $$
    
3. \textbf{Demanda máxima de lajotas} (2 mil):
    $$
    x_3 \leq 2
    $$
 

Adicionando folgas $s_1, s_2, s_3$:
 
\begin{align*}
x_1 + x_2 + x_3 + s_1 &= 12 \\
6x_1 + 8x_2 + 9x_3 + s_2 &= 90 \\
x_3 + s_3 &= 2
\end{align*}


\begin{array}{c|cccccc|c}
\text{Base} & x_1 & x_2 & x_3 & s_1 & s_2 & s_3 & b \\ \hline
s_1 & 1 & 1 & 1 & 1 & 0 & 0 & 12 \\
s_2 & 6 & 8 & 9 & 0 & 1 & 0 & 90 \\
s_3 & 0 & 0 & \mathbf{1} & 0 & 0 & 1 & 2 \\ \hline
Z   & -9 & -10 & \mathbf{-15} & 0 & 0 & 0 & 0
\end{array}


Variável que entra: $x_3$ (coeficiente $-15$, o mais negativo).
 
Teste da razão mínima:
$$
\frac{12}{1} = 12, \quad \frac{90}{9} = 10, \quad \frac{2}{1} = \mathbf{2} \leftarrow \text{mínimo}
$$
 
Variável que sai: $s_3$. Pivô $= 1$ (linha já normalizada).
 

Atualizando:
\begin{align*}
\text{Nova } s_1 &= s_1 - x_3: \quad (1,\ 1,\ 0,\ 1,\ 0,\ -1\ |\ 10) \\
\text{Nova } s_2 &= s_2 - 9x_3: \quad (6,\ 8,\ 0,\ 0,\ 1,\ -9\ |\ 72) \\
\text{Nova } Z   &= Z + 15x_3: \quad (-9,\ -10,\ 0,\ 0,\ 0,\ 15\ |\ 30)
\end{align*}
 

\begin{array}{c|cccccc|c}
\text{Base} & x_1 & x_2 & x_3 & s_1 & s_2 & s_3 & b \\ \hline
s_1 & 1 & 1 & 0 & 1 & 0 & -1 & 10 \\
s_2 & 6 & \mathbf{8} & 0 & 0 & 1 & -9 & 72 \\
x_3 & 0 & 0 & 1 & 0 & 0 & 1  & 2  \\ \hline
Z   & -9 & \mathbf{-10} & 0 & 0 & 0 & 15 & 30
\end{array}

 
 
Variável que entra: $x_2$ (coeficiente $-10$).
 

Teste da razão mínima (apenas linhas com coeficiente $> 0$ na coluna $x_2$):
$$
\frac{10}{1} = 10, \quad \frac{72}{8} = \mathbf{9} \leftarrow \text{mínimo}
$$
 
Variável que sai: $s_2$. Pivô $= 8$.
 

Nova linha $x_2 = \dfrac{1}{8}\, \text{linha}\, s_2$:
$$
\left(\frac{3}{4},\ 1,\ 0,\ 0,\ \frac{1}{8},\ -\frac{9}{8}\ \bigg|\ 9\right)
$$
 
Atualizando:
\begin{align*}
\text{Nova } s_1 &= s_1 - 1 \cdot x_2: \quad \left(\frac{1}{4},\ 0,\ 0,\ 1,\ -\frac{1}{8},\ \frac{1}{8}\ \bigg|\ 1\right) \\
\text{Nova } Z   &= Z + 10 \cdot x_2: \quad \left(-\frac{3}{2},\ 0,\ 0,\ 0,\ \frac{5}{4},\ \frac{15}{4}\ \bigg|\ 120\right)
\end{align*}
 

\begin{array}{c|cccccc|c}
\text{Base} & x_1 & x_2 & x_3 & s_1 & s_2 & s_3 & b \\ \hline
s_1 & \tfrac{1}{4} & 0 & 0 & 1 & -\tfrac{1}{8} & \tfrac{1}{8} & 1 \\
x_2 & \tfrac{3}{4} & 1 & 0 & 0 & \tfrac{1}{8}  & -\tfrac{9}{8} & 9 \\
x_3 & 0 & 0 & 1 & 0 & 0 & 1 & 2 \\ \hline
Z   & \mathbf{-\tfrac{3}{2}} & 0 & 0 & 0 & \tfrac{5}{4} & \tfrac{15}{4} & 120
\end{array}

 
Variável que entra: $x_1$ (único coeficiente negativo: $-\tfrac{3}{2}$).
 
Teste da razão mínima:
$$
\frac{1}{\,1/4\,} = \mathbf{4} \leftarrow \text{mínimo}, \quad \frac{9}{\,3/4\,} = 12
$$
 
Variável que sai: $s_1$. Pivô $= \tfrac{1}{4}$.
 

Nova linha $x_1 = 4 \times \text{linha}\, s_1$:
$$
\left(1,\ 0,\ 0,\ 4,\ -\frac{1}{2},\ \frac{1}{2}\ \bigg|\ 4\right)
$$
 
Atualizando:
\begin{align*}
\text{Nova } x_2 &= x_2 - \tfrac{3}{4} \cdot x_1: \quad \left(0,\ 1,\ 0,\ -3,\ \tfrac{1}{2},\ -\tfrac{3}{2}\ \bigg|\ 6\right) \\
\text{Nova } Z   &= Z + \tfrac{3}{2} \cdot x_1: \quad \left(0,\ 0,\ 0,\ 6,\ \tfrac{1}{2},\ \tfrac{9}{2}\ \bigg|\ 126\right)
\end{align*}
 

\begin{array}{c|cccccc|c}
\text{Base} & x_1 & x_2 & x_3 & s_1 & s_2 & s_3 & b \\ \hline
x_1 & 1 & 0 & 0 & 4  & -\tfrac{1}{2} & \tfrac{1}{2}  & 4 \\
x_2 & 0 & 1 & 0 & -3 & \tfrac{1}{2}  & -\tfrac{3}{2} & 6 \\
x_3 & 0 & 0 & 1 & 0  & 0             & 1             & 2 \\ \hline
Z   & 0 & 0 & 0 & 6  & \tfrac{1}{2}  & \tfrac{9}{2}  & \mathbf{126}
\end{array}

 
Todos os coeficientes na linha $Z$ são $\geq 0$: 
 

\begin{align*}
x_1 + x_2 + x_3 &= 4 + 6 + 2 = 12 \leq 12 \quad \checkmark \\
6(4) + 8(6) + 9(2) &= 24 + 48 + 18 = 90 \leq 90 \quad \checkmark \\
x_3 &= 2 \leq 2 \quad \checkmark
\end{align*}
 
$$
\boxed{x_1 = 4 \text{ mil blocos 6F}, \quad x_2 = 6 \text{ mil blocos 8F}, \quad x_3 = 2 \text{ mil lajotas}, \quad Z = R\$\ 126{,}00/\text{forno}}
$$
